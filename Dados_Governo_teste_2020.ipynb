# Gráfico Mês x Valor 
Podemos perceber que a partir de um cruzamento desses dois dados há um pico de gastos no final de 2018 para os parlamentares. Para inferir alguma conclusão para esses dados é necessário um cruzamento de maiores comportamentos dos parlamentares a partir dos dados cedidos. 

Ao utilizar plt.scatter  podemos visualizar de maneira simples esses dados. 

```python
fig,ax = plt.subplots()
x = df2['numMes']
y = df2['vlrLiquido']
plt.title('Mês X Valor') #adicionando o título
plt.scatter(x, y, marker="*", color='red')# definindo como será o gráfico e a cor
ax.annotate("Maior valor", 
            xy=(12, 120000),#coordenadas
            xycoords='data',
            xytext=(9, 90000),
            textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))# Anotação no gráfico
plt.show()
```

```python
fig,ax = plt.subplots()
x = df1['vlrDocumento']
y = df1['vlrLiquido']
plt.title('valorDocumento X Valor Liquído') #adicionando o título
ax.plot(x, y, color='green')
plt.show()
```
